```python
def splitTitle(sVal):
    return sVal.split(":")[0]

df["EmergencyCategory"] = df["title"].apply(splitTitle)
```

```python
# conver timestamp from object type to datetime type
df["timeStamp"] = pd.to_datetime(df["timeStamp"])
```
